```python
#Problem #7 Fahrenheit to Celsius
#https://www.codeabbey.com/index/task_view/fahrenheit-celsius


raw = [295, 335, 92, 589, 158, 182, 52, 503, 544, 352, 598, 499, 199, 286, 306, 148, 249, 465, 361, 59, 422, 211, 512, 418, 376, 491, 167, 167, 543, 379, 134, 237, 113, 195, 226, 240, 345, 247, 142, 289]

def converter(an_input):
    
    celsius = []
    celsius_string = ''
    
    for number in an_input:
        celsius.append(round(((number - 32)*5)/9))
        
    for temp in celsius:
        celsius_string += str(temp) + ' ' 
        
    return celsius_string
```
